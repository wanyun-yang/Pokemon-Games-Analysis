```python
from bs4 import BeautifulSoup
import requests
from numpy import nan as NA
import numpy as np
import pandas as pd
pd.set_option('max_colwidth',150)
```

```python
pokemon_url = 'https://pokemondb.net/pokedex/all'
# Issue a simple HTTP request to get the webpage text
pokemon_page = requests.get(pokemon_url)
# Response code is returned
pokemon_page
```

```python
pokemon_page_soup = BeautifulSoup(pokemon_page.text, 'html.parser')
```

```python
# Define a function that takes in a row of the pokedex table and returns it as a DataFrame with a single row. 
def pokemon_df(number):
    if number<=0:
        print('This pokemon does not exist.')
    else:
        # Extract the pokemon's name from the web
        pokemon=pokemon_page_soup.find_all(name="tr")[number]
        pokemon_values_list = pokemon.find_all(name="td")

        name_str = pokemon_values_list[1].text

        # Extract the pokemon's individual url address from the web
        url_str = pokemon.find(name="a", class_="ent-name").get('href')
        url_str = pokemon_url[0:21] + url_str


        # Extact the type(s) of the pokemon 
        type_str = pokemon_values_list[2].text

        # Extract the total points of the pokemon
        total_str = pokemon_values_list[3].text

        # The list of ID Number, HP, Attack, Defense, Sp. Atk, Sp. Def, Speed
        num_list=[]
        for i in range(7):
            num_list.append(pokemon.find_all(name="td", class_="cell-num")[i].text)
        # Return result as a dataframe created from a NP array
        pokemon_row = np.array([[ num_list[0],name_str, url_str, type_str, total_str,num_list[1],num_list[2],num_list[3],num_list[4],num_list[5],num_list[6]]])
        pokemon_df = pd.DataFrame(pokemon_row,columns = ['ID','NAME', 'URL', 'TYPE(s)', 'TOTAL', 'HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed'])
        return pokemon_df
```

```python
pokemon_ids = [i+1 for i in range(1028)]
# Loop over our route IDs and append to the empty dataframe
# Store dataframes in a list
dfs = []
for id in pokemon_ids:
    pokemon_scrape_res = pokemon_df(id)
    # Append df to list
    dfs.append(pokemon_scrape_res)
```

```python
pokemon_scrape_df = pd.concat(dfs)
```

```python
# I have added the column names
# Convert str to numericals
pokemon_scrape_df.loc[:,'ID'] = pokemon_scrape_df.ID.astype(int)
pokemon_scrape_df.loc[:,'TOTAL'] = pokemon_scrape_df.TOTAL.astype(int)
pokemon_scrape_df.loc[:,'HP'] = pokemon_scrape_df.HP.astype(int)
pokemon_scrape_df.loc[:,'Attack'] = pokemon_scrape_df.Attack.astype(int)
pokemon_scrape_df.loc[:,'Defense'] = pokemon_scrape_df.Defense.astype(int)
pokemon_scrape_df.loc[:,'Sp_Atk'] = pokemon_scrape_df.Sp_Atk.astype(int)
pokemon_scrape_df.loc[:,'Sp_Def'] = pokemon_scrape_df.Sp_Def.astype(int)
pokemon_scrape_df.loc[:,'Speed'] = pokemon_scrape_df.Speed.astype(int)
```

```python
# Remove duplicate values of pokemon based on the URL. 
result_df = pokemon_scrape_df.drop_duplicates(subset=['URL'], keep='first')
```

```python
# Define a function that makes in a single pokemon's location table and returns it as a DataFrame with a single row. 
def pokemonLocation_df(ID):
    # Extract the pokemon's name from the web
    pokemon_url =result_df.iat[ID-1, 1]
    # Extract the location table
    tables = pd.read_html(requests.get(pokemon_url, headers={'User-agent': 'Mozilla/5.0'}).text)
    
    if ID >=807:
        loc_df = tables[-1]
    else:
        loc_df = tables[-2]
        
    loc_df.columns = ["GAMES", "PLACE"]
    loc_df = loc_df.set_index(['GAMES'])
    loc_df = loc_df.transpose()
    
    gamesOrigin=loc_df.columns.tolist()
    
    gamesNames = ['Red', 'Blue', 'Yellow', 'Gold', 'Silver', 'Crystal', 'Ruby', 'Sapphire', 'FireRed', 'LeafGreen', 'Emerald','Diamond','Pearl','Platinum','HeartGold','SoulSilver','Black','White','Black 2','White 2','X','Y','Omega Ruby','Alpha Sapphire','Sun','Moon','Ultra Sun','Ultra Moon',"Let's Go Pikachu","Let's Go Eevee",'Sword','Shield']
    content=[]
    for j in range(len(gamesNames)):
        i=0
        while i>=0:
            if gamesOrigin[i].find(gamesNames[j])!=-1:
                # When I search for Y game version, the index will goes to 1, since 'Yellow' has a 'Y' too, so I take it as a special case.
                if j==21 and gamesOrigin[i]=='Yellow':
                    i+=1
                else:
                    content.append(loc_df.iat[0,i])
                    break;
            else:
                i=i+1
            if i==len(gamesOrigin):
                content.append(np.nan)
                break;

    location = pd.DataFrame(content).transpose()
    location.columns = gamesNames
    location.insert(0, "NAME", result_df.iat[ID-1, 0], True)
    location.insert(0, "ID", result_df.index[ID-1], True)
    # Return result as a dataframe created from a NP array
    return location
```

```python
# Loop over our sample IDs and append to the empty dataframe
dfs = []
for id in sample1.index:
    pokemon_scrape_res = pokemonLocation_df(id)
    #Append df to list
    dfs.append(pokemon_scrape_res)
pokemonLocations= pd.concat(dfs)
pokemonLocations = pokemonLocations.set_index(['ID'])
pokemonLocations.head()
```

# Exercise #3 - Location Analysis Done Right

### STEP 3.1

```python
# Join the pokedex data to the  location DataFrame
pokemonStat=pokemonLocations.reset_index().merge(result_df, how="left").set_index(pokemonLocations.index.names)
pokemonStat
```

```python
pokemonX=pokemonStat[['NAME','X','TOTAL']]
pokemonX=pokemonX[pokemonX.index<= 720]
pokemonX
```

```python
def explode_str(df, col, sep):
    s = df[col]
    i = np.arange(len(s)).repeat(s.str.count(sep) + 1)
    return df.iloc[i].assign(**{col: sep.join(s).split(sep)})
```

```python
# separate different locations in one cell
pokemonX=explode_str(pokemonX, 'X', ',')
pokemonX
```

```python
#For the locations in pokemon X, calculate the average total points for each location. 
pokemonX.groupby(by='X')['TOTAL'].mean().sort_values(ascending=False)
```
